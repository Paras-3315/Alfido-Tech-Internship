# Loading and cleaning the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('traffic.csv')
print(df)
```

```python
#Data cleaning and preprocessing
```

```python
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
print(df.isnull().sum())
```

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.duplicated())
```

```python
df["date"] = pd.to_datetime(df['date'],dayfirst = True)
```

```python
print(df["date"])
```

```python
df.dropna(subset=['date'], inplace = True)
```

```python
print(df["date"])
```

```python
print(df.columns)
```

# Traffic Analysis

```python
#1. Traffic trends over time
trend = df.groupby('date').size()
print(trend)
```

```python
plt.figure(figsize=(10,5))

plt.plot(trend.index,trend.values,marker='o',linestyle='-',color="green",mfc="red",mec="red",ms=10)
plt.xlabel("Date")
plt.ylabel("Number of events")
plt.title("Website Traffic Analysis")
plt.show()
```

```python
#2. User behaviour analysis
freq_track = df["track"].value_counts().head(10)
freq_album = df["album"].value_counts().head(10)
freq_artist = df["artist"].value_counts().head(10)
print(freq_track)
```

```python
plt.figure(figsize=(15,8))
plt.subplot(2,3,1)
freq_track.plot(kind="barh",color="blue")
plt.title("Most frequently accessed tracks")

plt.subplot(2,3,3)
freq_album.plot(kind="barh",color="green")
plt.title("Most frequently accessed albums")

plt.subplot(2,3,5)
freq_artist.plot(kind="barh",color="red")
plt.title("Most frequently accessed artists")
plt.show()
```

```python
#3. Geographic analysis
```

```python
country = df["country"].value_counts().head(10)
city = df["city"].value_counts().head(10)
print(country)
print(city)
```

```python
plt.figure(figsize=(10,8))

plt.suptitle("Geographic Analysis")
plt.subplot(2,1,1)
plt.barh(country.index,country.values,edgecolor="black")
plt.title("Top 10 countries by traffic")

plt.subplot(2,1,2)
plt.barh(city.index,city.values,color="green",edgecolor="black")
plt.title("Top 10 cities by traffic")
plt.show()
```

```python
#4. Sources and conversion analysis
```

```python
ref = df["linkid"].value_counts(normalize=True).head(10)*100
source = df["isrc"].value_counts(normalize=True).head(10)*100
print(ref)
print(source)
```

```python
plt.figure(figsize=(10,8))

plt.suptitle("Source & Conversion Analysis (in %)")
plt.subplot(2,1,1)
plt.barh(ref.index,ref.values,edgecolor="black")
plt.title("Top 10 traffic sources")

plt.subplot(2,1,2)
plt.barh(source.index,source.values,color="green",edgecolor="black")
plt.title("Top 10 tracks by ISRC")
plt.show()
```
